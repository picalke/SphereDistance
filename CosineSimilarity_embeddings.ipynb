```python
import scipy as sp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import openai
from sklearn.datasets import fetch_20newsgroups, fetch_20newsgroups_vectorized
import random
```

```python
batch = fetch_20newsgroups()
```

```python
#Random sentences
def choose_rnd_sent(text, n=1, min_length=20, max_length=150):
    sentences = text.split(".")
    # Remove empty sentences (if any)
    sentences = [sentence.strip() for sentence in sentences if sentence.strip()]
    random_sentences = []
    for _ in range(n):
        success = False
        for _ in range(100):
            random_sentence = random.choice(sentences)
            if min_length <= len(random_sentence) <= max_length:
                random_sentences.append(random_sentence)
                success = True
                break
        if not success:
            break
    return random_sentences
rnd_msgs=[ random.choice(batch.data) for i in range(1500) ]
rnd_sents= [(choose_rnd_sent(i))[0]  for i in rnd_msgs ]
```

```python
rnd_flat=[]
for k in rnd_sents:
    if len(k)==1:
        rnd_flat.append(k[0])
```
